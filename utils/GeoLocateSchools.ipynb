```python
import urllib
from urllib.request import Request, urlopen

import time
```

```python
url = 'https://www.usnews.com/education/best-high-schools/search?format=json&district-id=110570&state-urlname=illinois&page={}'
```

```python
max_attempts = 3
attempts = 0
sleeptime = 10 #in seconds, no reason to continuously try if network is down
num_pages = 30


for page in range(1,num_pages+1):
    #while true: #Possibly Dangerous
    while attempts < max_attempts:
        try:
            print(url.format(page))
            req = Request(url.format(page), 
                          headers={'User-Agent': 'Mozilla/5.0'})

            with urlopen(req) as response:
                content = response.read()
            print(content)
            break
        except Exception as e:
            attempts += 1
            print(e)
        time.sleep(sleeptime)
    break
```

```python
import json
from os import listdir
import os
import pandas as pd
import numpy as np
```

```python
columns = set()
schools = []
    
root_dir = '../data/schools'
for name in listdir(root_dir):
    print('name:',name)
    if '.json' in name:
        with open(os.path.join(root_dir,name)) as f:
            data = json.load(f)
        for result in data['matches']:
            try:
                school = {
                    'name':result['name'],
                    'rank':result['ranks']['primary_rank']['value']
                }
                schools.append(school)
                for data in result['data']:
                    columns.add(data['label'])
                    if isinstance(data['raw_value'], str):
                        school[data['label']] = int(data['raw_value'].replace(',',''))
                    else:
                        school[data['label']] = data['raw_value']
            except Exception as e:
                continue
```

```python
columns = ['rank','name'] + list(columns)
```
